### 2.1 : <u>Assigning Correct Datatype</u>
_Changing Datatype of **`Request id`** and **`Driver id`** to **Object Datatype** as they are categorical variables._

```python
# Changing Datatype of `Request id` and `Driver id` to Object Datatype
uber_df['Request id'] = uber_df['Request id'].astype(object)
uber_df['Driver id']  = uber_df['Driver id'].astype(object)
```

_Now, changing Datatype of **`Request timestamp`** and **`Drop timestamp`** to **Datetime Datatype**._<br>
_For the same we need to create a new function, as our date-time data are in two different format. Thus we need to convert and change multiple format into one standard Datetime format._

```python
# Creating new function to correct the date format of the data and change its Datatype to Datetime Datatype.

def correct_date_format(date_str):
    try:
        return pd.to_datetime(date_str, format='%d/%m/%Y %H:%M', errors='raise')  # First, Tries to parse dates as Day/Month/Year
    except ValueError:
        try:
            return pd.to_datetime(date_str, format='%m/%d/%Y %H:%M', errors='raise') # Secondly, Tries to parse dates as Month/Day/Year if the first fails.
        except ValueError:
            try:
                return pd.to_datetime(date_str, format='%d-%m-%Y %H:%M:%S', errors='raise') # lastly, handles formats that use hyphens as Day-Month-Year
            except ValueError:
                return pd.to_datetime(date_str, errors='coerce')
```

```python
# Applying the Defined Function to changing datatype of `Request timestamp` and `Drop timestamp` to Datetime Datatype.  

uber_df['Request timestamp'] = uber_df['Request timestamp'].apply(correct_date_format)
uber_df['Drop timestamp'] = uber_df['Drop timestamp'].apply(correct_date_format)
```

```python
# Now, Checking Info for Changes
uber_df.info()
```

### 2.2 : <u>Removing Case Mis-match

```python
# Again, creating New Function to convert alphabetic characters to uppercase
def to_uppercase_alpha(val):
    if isinstance(val, str):
        return ''.join([char.upper() if char.isalpha() else char for char in val])
    return val
```

```python
# Removing case mis-match by Standardizing all the values of string type to Uppercase() using the Function Created.

uber_df = uber_df.map(lambda x: to_uppercase_alpha(x) if isinstance(x, str) else x)
```

### 3.1 : <u>Extracting Date & Hour from **`Request timestamp`** and storing in new columns **`Request_date`** and **`Request_hour`**

```python
# Creating new columns 'Request_date' and 'Request_time' to store date and time from `Request timestamp` respectively.

uber_df['Request_date'] = pd.to_datetime(uber_df['Request timestamp'].dt.date)
uber_df['Request_hour'] = uber_df['Request timestamp'].dt.hour
uber_df.head(5)
```

### 3.2 : <u>Let's divide the Hours of a Day into different zones or time slots and storing them in new variable **`Request_timeslot`**

```python
# Creating a new function to categorize timeslot based upon hour of the day.

def categorize_timeslots(hour):
    if 0 <= hour < 3:            # 12AM - 3AM : Late Night
        return 'LATE NIGHT'
    elif 3 <= hour < 7:          # 3AM - 7AM : Early Morning
        return 'EARLY MORNING'
    elif 7 <= hour < 12:         # 7AM - 12PM : Morning
        return 'MORNING'         
    elif 12 <= hour < 16:        # 12PM - 4PM : Afternoon
        return 'AFTERNOON'
    elif 16 <= hour < 20:        # 4PM - 8PM : Evening
        return 'EVENING'
    else:                        # 8PM - 12AM : Night
        return 'NIGHT'
```

```python
# Applying function to create new column 'Request_timeslot'
uber_df['Request_timeslot'] = uber_df['Request_hour'].apply(categorize_timeslots)
```

### 3.3 : <u>Droping **`Drop timestamp`** , as it's not required for our Analysis.


```python
# Droping 'Drop timestamp'
uber_df.drop('Drop timestamp', axis=1 , inplace=True)
```

### 5.1 : <u>Now, let's create new columns for ride **`Demand`**, ride **`Supply`** and for ride **`Gap`**.

```python
# Creating Supply Column, (when Ride request gets completed, i.e. when ride 'Status' is 'TRIP COMPLETED')

uber_df.loc[uber_df['Status']=='TRIP COMPLETED', 'Supply'] = 'Supply'
```

```python
# Creating Demand Column (when Ride is requested, i.e. when ride `Request id` is not Null)

uber_df.loc[~uber_df['Request id'].isnull(), 'Demand'] = 'Demand'
```

```python
# Creating Gap Column, (when ride request are not completed, i.e. when ride `Status` is eihter `Cancelled` or `No Cars Available`)

uber_df.loc[(uber_df.Status == 'CANCELLED') | (uber_df.Status == 'NO CARS AVAILABLE'), 'Gap'] = 'Gap' 
```
